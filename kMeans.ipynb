```python
import seaborn as sns 
from sklearn.datasets import make_blobs # generate a clustering data  
```

```python
X , y = make_blobs(
    n_samples=1000,
    n_features=2,
    centers= 4 , # clusters 
    random_state=42
)
```

```python
y # showing the true value 
```

```python
# Visulaize 

sns.scatterplot(x=X[:,0] , y=X[:,1] ) 
```

```python
# k -means 

from sklearn.cluster import KMeans
```

```python
k = 4 

kmeans =KMeans(
    n_clusters=k,
    init= "k-means++" ,
    random_state=42
    
)
```

```python
labels = kmeans.fit_predict(X) # lables cluster  number 
```

```python
labels
```

```python
sns.scatterplot(x=X[:,0] , y=X[:,1] , c=labels)
```

### How to Choose K value - Elbow Method ,  Silhouette Score 

```python
# Elbow Method 
wcss = []
for k in range(1,21):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit_predict(X)
    wcss.append(kmeans.inertia_)
```

```python
sns.lineplot(x=range(1,21) , y=wcss , marker="o")
```

# Silhouette score  

```python
from sklearn.metrics import  silhouette_score  
```

```python
ss = []

for k in range(2,21):  # silhouette score is always start with 2 
    kmeans = KMeans(n_clusters=k)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X , labels)
    ss.append(score)
    
```

```python
# create a plot 
```

```python
sns.lineplot(x=range(2,21) , y=ss , marker="o")
```

```python
# who's silhouette score value is maximum  is always good 
```
